```kotlin
val years by columnOf("2018", "2019", "2020", "2021", "2022")
val cost by columnOf(62.7, 64.7, 72.1, 73.7, 68.5)
val df_year_costs = dataFrameOf(years, cost)

df_year_costs.plot {
    line {
        x(years)
        y(cost)
    }
}
```
